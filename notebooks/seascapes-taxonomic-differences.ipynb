### grouped bar chart - `plotly`

```python
_df = pd.DataFrame(columns = ['class', 'taxa', 'value'])
_df
```

```python
for i in df.iterrows():
    i[1].dropna(inplace=True)
    if(len(i[1])) == 0:
        continue
    temp = pd.DataFrame({
        'class': i[0],
        'taxa': i[1].index,
        'value': i[1].values
    })
    _df = pd.concat([_df, temp])
```

```python
_df.index = range(len(_df))
```

```python
_df
```

```python
_df.value = _df.value.astype(float)
_df["class"] = _df["class"].astype(str)
```

```python
_df.sort_values(by=["class", "value"], ascending=[True, False], inplace=True)
```

```python
fig = px.bar(_df, 
             x = "class",
             y = 'value',
             text = _df['taxa'] + " " + _df['value'].round(1).astype(str) + "%",
             color = 'value',
             labels = {
                 'value':'percentage (%)',
                 'class':'seascape class',
             },
             color_continuous_scale = "teal",
            title="Taxanomic distribution v/s Seascape Class")
fig.show()
```

### What if we remove the dominant taxa and see the less abundant species?

```python
fig = px.bar(_df[_df["taxa"]!='Actinopteri'], 
             x = "class",
             y = 'value',
             text = _df[_df["taxa"]!='Actinopteri']['taxa'] + " " + _df[_df["taxa"]!='Actinopteri']['value'].round(1).astype(str) + "%",
             color = 'value',
             labels = {
                 'value':'percentage (%)',
                 'class':'seascape class',
             },
             color_continuous_scale = "teal",
            title="Taxonomic distribution other than Actinopteri v/s Seascape Class")
fig.show()
```

```python
print("\t\t\t\t\t\t\t\tTaxonomic distribution other than Actinopteri v/s Seascape Class")
print("\t\t\t\t\t\t\t\t================================================================")
fig, ax = plt.subplots(2, 3)
j, k = 0, 0
fig.set_size_inches(20,8)
fig.set_dpi(300)
for i in _df[_df["taxa"]!='Actinopteri']["class"].unique():
    sns.barplot(data=_df[_df["taxa"]!='Actinopteri'][_df[_df["taxa"]!='Actinopteri']["class"]==i], x = "value", y="taxa", ax = ax[j][k])
    
    ax[j][k].set_title(f"Class: {i}")
    ax[j][k].set_xlabel("percentage (%)")
    if k==2:
        j = 1
    k = (k+1)%3
    fig.tight_layout(pad=1)
plt.show()
```
